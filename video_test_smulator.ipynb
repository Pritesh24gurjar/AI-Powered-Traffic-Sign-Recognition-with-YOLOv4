```python
# img_file = './data/obj.names'
# classNames = ['crosswalk', 'speedlimit', 'stop', 'trafficlight']
classNames = ['prohibitory','danger','mandatory','other']
print("Classes' names :\n", classNames)
```

```python
def danger():
   mixer.init()
   alert=mixer.Sound('D:\Drowsiness-monitoring-Using-OpenCV-Python-master\Audio\danger.wav')
   alert.play()
   time.sleep(0.3)
#    alert.play()  
def prohibitory():
   mixer.init()
   alert=mixer.Sound('D:\Drowsiness-monitoring-Using-OpenCV-Python-master\Audio\prohibitory.wav')
   alert.play()
   time.sleep(0.3)
#    alert.play()
def mandatory():
   mixer.init()
   alert=mixer.Sound('D:\Drowsiness-monitoring-Using-OpenCV-Python-master\Audio\mandatory.wav')
   alert.play()
   time.sleep(0.3)
#    alert.play()
def other():
   mixer.init()
   alert=mixer.Sound('D:\Drowsiness-monitoring-Using-OpenCV-Python-master\Audio\other.wav')
   alert.play()
   time.sleep(0.3)
#    alert.play()
```

```python
modelConfig_path = './weights/yolov3_training.cfg'
modelWeights_path = './weights/yolov3_training_last.weights'
# read the model cfg and weights with the cv2 DNN module
neural_net = cv2.dnn.readNetFromDarknet(modelConfig_path, modelWeights_path)
# set the preferable Backend to GPU
neural_net.setPreferableBackend(cv2.dnn.DNN_BACKEND_CUDA)
neural_net.setPreferableTarget(cv2.dnn.DNN_TARGET_CUDA)
```

```python
network = neural_net
height, width = 416,416

# confidence and non-max suppression threshold for this YoloV3 version
confidenceThreshold = 0.3
nmsThreshold = 0.6
capture =cv2.VideoCapture(0)
while True:
    ret, img = capture.read()
    # gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    # faces = face_cascade.detectMultiScale(gray)
    outputs = convert_to_blob(img, network, height, width)    

    bounding_boxes, class_objects, confidence_probs = object_detection(outputs, img, confidenceThreshold)  
    indices = nms_bbox(bounding_boxes, confidence_probs, confidenceThreshold, nmsThreshold)

    temp , class_lbl=box_drawing(img, indices, bounding_boxes, class_objects, confidence_probs, classNames, color=(	255, 0, 0), thickness=2)

    class_lbl = np.array(class_lbl)
    class_lbl = np.unique(class_lbl)
    for i in class_lbl:
        print(i)
        if(i=="danger"):
            danger()
        elif(i=="prohibitory"):
            prohibitory()
        elif(i=="mandatory"):
            mandatory()
        else:
            other()
        
    cv2.imshow('frame',img)

    if cv2.waitKey(1) & 0xff == ord('q'):
        break
capture.release()
cv2.destroyAllWindows()
##########################################
# load the image
# img = cv2.imread('images/00894.jpg')

# cv2.imshow('Image',img)
# # cv2.imshow('image', img)
# # using convert_to_blob function : 
# outputs = convert_to_blob(img, network, height, width)    
# # apply object detection on the video file
# bounding_boxes, class_objects, confidence_probs = object_detection(outputs, img, confidenceThreshold)  
# # print(outputs)
# # print(bounding_boxes)
# # print(class_objects)
# # print(confidence_probs)
# # perform non-max suppression
# indices = nms_bbox(bounding_boxes, confidence_probs, confidenceThreshold, nmsThreshold)
# # draw the boxes
# temp=box_drawing(img, indices, bounding_boxes, class_objects, confidence_probs, classNames, color=(255,255,0), thickness=2)

# print(temp)
# # to save the detected image
# img_save = cv2.imwrite('./output/yolov4res19.jpg', img)

# cv2.imshow('Object detection in images', img)
# cv2.waitKey()
# cv2.destroyAllWindows()
```
